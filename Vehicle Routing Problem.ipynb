```python
#VEHICLE ROUTING PROBLEM 

#FORMULACION MATEMATICA DEL PROBLEMA 


# CLIENTES (C) : CONJUNTO DE CLIENTES , EXEPTO EL CENTRO DE DISTRIBUCION 
# NODOS : CONJUNTO DE NODOS QUE ES LA UNION DE EL CONJUNTO DE CLIENTES MÃS EL CENTRO DE DISTRIBUCION (DEPOT)
# ARCOS : CONJUNTO DE TODOS LOS ARCOS DE CADA NODO DONDE I SEA DISTINTO DE J 
# COSTO (Cij) : COSTO O DISTANCIA ENTRE NODO I Y NODO J 
# VOLUMEN (Uij): CARGA DE CADA PRODUCTO DEL CLIENTE 
# CAPACIDAD (Q) : CAPACIDAD DE LOS VEHICULOS 


#VARIABLES DE DESICION

# Xij = 1 , cuando se decide viajar en el arco (i,j)-> A , cero en caso contrario 
# Ci = capacidad de cada cliente / es decir que cada cliente tiene una capacidad que se debe satisfacer 


#FUNCION OBJETIVO 

# MIN Z = COSTO TOTALES

# RESTRICCIONES 

#1 Y 2 : LA SUMATORIA DE TODAS LAS BINARIAS DEBE SER IGUAL A 1 ,  ES DECIR QUE CADA NODO DEBE VISITAR SOLO UN NODO 

        # Y TAMBIEN QUE TODOS LOS NODOS SOLO PUEDEN SER VISITADOS POR UN SOLO NODO


# RESTRICCIONES DE CAPACIDAD : RESTRINGIR LA CAPACIDAD DEL VEHICULO
                                
                             
# Lectura de las instancias de CVRPLIB

#A-n32-k5

# PRIMERO DEBEMOS LEER DEL ARCHIVO 

import numpy as np 

# Nombre de la instancia 
fname = 'Augerat/A-n32-k5.vrp'

#CREAMOS UNA VARIABLE LINEAS COMO LISTA PARA ALMACENAR LA DATA DE LA INSTANCIA

lineas = []

#ABRIMOS LA INSTANCIA Y RECOPILAMOS LA DATA COMO LISTA

with open (fname , 'rt') as f:
    for f_lineas in f:
        lineas.append(f_lineas)

        
print(lineas)
#OBTENIENDO CAPACIDAD DE LOS VEHICULOS , DIMENSION(CANTIDAD DE NODOS) , CANTIDAD DE VEHICULOS COMO ENTEROS 

dimension_str = lineas[3]
capacidad_str = lineas[5]
vehiculos_str = lineas[0]

#CAMBIANDO EL TIPO DE STRING A INTEGER

capacidad = int((capacidad_str[10:]))

cantidad_nodos = int(dimension_str[12:])

vehiculos = int(vehiculos_str[14:])


print("La cantidad de vehiculos son de:" , vehiculos)
print("\n")
print("La capacidad de los vehiculos es de : " ,  capacidad)
print("\n")
print ("La cantidad de nodos que tiene la instancia es: " , cantidad_nodos)
print("\n")


extraer_coordenadas = cantidad_nodos + 7


# OBTENGO LOS DATOS DE LAS COORDENADAS DE LOS NODOS EN UN ARREGLO 

print ("Esta es la Lista de Coordenadas\n")

data_coordenadas = lineas[7 : extraer_coordenadas]

print (data_coordenadas)

print("\n")

# AHORA HAY QUE OBTENER LAS DEMANDAS DE CADA CIUDAD 

print("Esta es la lista de Demandas\n")

extraer_demandas = extraer_coordenadas + 1

data_demandas = lineas[extraer_demandas : extraer_demandas + cantidad_nodos ]

print (data_demandas)

print("\n")


# SEPARAR LOS DATOS DE EL ARREGLO DE COORDENADAS 

limpieza_coordenadas = []
limpieza_demandas = []

for x in data_coordenadas:
    limpieza_coordenadas.append(x.split())
    
for y in data_demandas:
    limpieza_demandas.append(y.split())
    

print("Coordenadas Limpias")
print(limpieza_coordenadas)
print("\n")
print("Demandas Limpias")
#print(limpieza_demandas)
print("\n")

# SEPERAR POS_X , POS_Y Y CALCULAR LA DISTANCIA ENTRE LOS NODOS 

print(limpieza_coordenadas)

distancia = {(i,j):np.hypot(int(limpieza_coordenadas[i][1]) -  int(limpieza_coordenadas[j][1]) ,  int(limpieza_coordenadas[i][2]) -  int(limpieza_coordenadas[i][2]) ) for i in range(0, cantidad_nodos ) for j in range(0,cantidad_nodos) if i!=j }

distancia

    







    
    
    
 
    




```

```python
# CREACION DE LOS DATOS 

#Libreria numpy para la funcion random 



#Cantidad de clientes
n = 10

#Lista de los clientes 

clientes = [x for x in range(1,n+1)] # RECORDAR QUE EL 0 ES EL DEPOSITO O DEPOT

# NODOS ES LA UNION ENTRE DEPOT + LISTA CLIENTES 

nodos = [0] + clientes 

Q = 15 # CAPACIDAD DEL VEHICULO 

rnd = np.random

rnd.seed(1)


# CREO UN DICCIONARIO CON LA LISTA DE CLIENTES COMO CLAVE Y LA CAPACIDAD DE CARGA COMO DATO 
                                                                
q = {i : rnd.randint(1,10) for i in clientes} # CICLO PARA ASIGNAR A CADA CLIENTE UNA CARGA RANDOM


```

```python
#CREAR COORDENADAS 
rnd.seed(1)
loc_x = rnd.rand(len(nodos)) * 200 # 
loc_y = rnd.rand(len(nodos)) * 100
```

```python
#GRAFICO DE LOS NODOS 

import matplotlib.pyplot as plt

plt.figure(figsize = (12 , 5)) # TAMAÃ‘O DEL CANVAS O DEL GRAFICO 
plt.scatter(loc_x , loc_y , color = "black") # GRAFICO DE DISPERSION DE LOS NODOS 

for i in clientes :
    plt.annotate('$q_{%d}=%d$'%(i , q[i]) , (loc_x[i]+1 , loc_y[i]-0.5) ) #PARA MOSTRAR LOS NODOS EN EL GRAFICO 

    
plt.plot(loc_x[0] , loc_y[0] , color="red" , marker='s')    
plt.annotate("DP" , (loc_x[0] , loc_y[0]+2))
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("GrÃ¡fico de nodos | VRP")


plt.show()









```

```python
#Creando Estructuras de datos (Arcos y Distancias entre arcos ) 

arcos = {(i,j) for i in nodos for j in nodos if i!=j  }

distancia = {(i,j):np.hypot(loc_x[i] - loc_x[j] ,  loc_y[i] - loc_y[j] ) for i in nodos for j in nodos if i!=j }

distancia
```

```python
#CREACION DEL MODELO 

# EN ANACONDA PROMP COLOCAR PIP INSTALL DOCPLEX (listo)

from docplex.mp.model import Model
mdl = Model('VRP')
```

```python
#VARIABLES DE DESICION

x = mdl.binary_var_dict(arcos , name="x")

#ub = upperbound , quiere decir que u no puede ser mayor que la capacidad del vehiculo
u = mdl.continuous_var_dict(nodos , ub=Q, name="u" ) 

```

```python
#FUNCION OBJETIVO

mdl.minimize(mdl.sum(distancia[i,j] * x[i,j] for i,j in arcos ))
```

```python
#RESTRICCIONES NODOS

mdl.add_constraints(mdl.sum(x[i,j] for j in nodos if i!=j)==1 for i in clientes )
mdl.add_constraints(mdl.sum(x[i,j] for i in nodos if i!=j)==1 for j in clientes )

#RESTRICCIONES DE INDICACION

mdl.add_indicator_constraints(mdl.indicator_constraint(x[i,j] , u[i] + q[j] == u[j])
                             for i,j in arcos if i!=0 and j!=0)


mdl.add_constraints(u[i]>=q[i] for i in clientes)


#print(mdl.export_to_string())


```

```python

mdl.parameters.timelimit=120
solucion=mdl.solve(log_output=True)

```
